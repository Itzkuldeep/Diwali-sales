## EDA

### Gender

```python
ax = sns.countplot(x = 'Gender', data = df)
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

gx = sns.barplot(x ='Gender', y ='Amount', data  = sales_gen)
```

## AGE

```python
sns.countplot(data = df, x = 'Age Group')
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.barplot(data = sales_age, x = "Age Group", y = 'Amount')
```

## State

```python
df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(15)
sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(15)
sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
```

## Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status' )

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_ms = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending =False)
sns.set(rc = {'figure.figsize': (7,5)})
sns.barplot(data = sales_ms, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

```python
sns.set(rc = {'figure.figsize' : (20,5)})
ax = sns.countplot(data = df, x = 'Occupation' )

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_oc = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(data = sales_oc, x = 'Occupation', y = 'Amount')
```

*From above we can say that most buyers are working in IT sector*

## Product Category

```python
sns.set(rc = {'figure.figsize' : (22,5)})
ax = sns.countplot(data = df, x = 'Product_Category' )

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_oc = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sns.set(rc ={'figure.figsize':(25,5)})
sns.barplot(data = sales_oc, x = 'Product_Category', y = 'Amount')
```

```python
sales_pd = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc ={'figure.figsize':(25,5)})
sns.barplot(data = sales_pd, x = 'Product_ID', y = 'Orders')
```
